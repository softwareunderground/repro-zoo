# Figure 1

Authors start by introducting a wavelet *R<sub>1</sub>*, its time-shifted variant *R<sub>2</sub>* in Figure 1.a. Then they show the difference *R<sub>d</sub>* in Figure 1.b., and finally the ray paths for *R<sub>1</sub>* and *R<sub>2</sub>* in Figure 1.c.

It is not explicitly specified what the wavelet is, however, it looks like a Gaussian modulated sinusoid wavelet. We will make the first figure with slight improvements in anootations; and roughly estimating some of the other things.

```python
import numpy as np
from scipy.signal import gausspulse

from figures import figure1

shift = 0.025

time_start = 0  # seconds
time_stop = 1  # seconds
time_step = 0.01  # seconds

wavelet_freq = 3  # Hz

# Make inclusive range [start, stop]
times = np.arange(time_start, time_stop + time_step, time_step) - 0.5

# Make R_1 R_2 and R_d
amplitude_r1 = -1 * gausspulse(t=times, fc=wavelet_freq)
amplitude_r2 = -1 * gausspulse(t=times - shift, fc=wavelet_freq)
amplitude_rd = amplitude_r1 - amplitude_r2

# Plot
figure1(times, amplitude_r1, amplitude_r2, amplitude_rd, shift, dpi=130)
```
